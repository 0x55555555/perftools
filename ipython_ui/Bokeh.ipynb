```python
from pymongo import MongoClient
client = MongoClient('localhost', 27017)
perf = client.perf.perf_data
```

```python
data = perf.aggregate([
    {
        "$group": {
            "_id": { "desc": "$recipeDescription", "branch": "$branch", "machineIdentity": "$contexts.test.machineIdentity" },
            "data": { "$push": "$contexts.test"}
        }
    }
])['result'][0]

results = data['data']

result_data = [x['results'] for x in results]

result_data
```

```python
from numpy.random import beta
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
%matplotlib inline  

figsize(16.5, 9)

plt.style.use('fivethirtyeight')


def plot_beta_hist(a, b):
    plt.hist(beta(a, b, size=10000), histtype="stepfilled",
             bins=25, alpha=0.8, normed=True)
    return

plot_beta_hist(10, 10)
plot_beta_hist(4, 12)
plot_beta_hist(50, 12)
plot_beta_hist(6, 55)

plt.show()
```
